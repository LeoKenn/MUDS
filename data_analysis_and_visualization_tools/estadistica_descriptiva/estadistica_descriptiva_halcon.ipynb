###Representar un histograma con la distribución de la población de la ciudad por edades

```r

library(dplyr)

# Modificar el conjunto de datos para que empiece en la tercera fila y quitar las primeras tres columnas
datos_modificados <- data %>%
  slice(2:n()) %>%    # Seleccionar desde la tercera fila hasta la última
  select(-1:-3)       # Excluir las primeras tres columnas

# Verificar el conjunto de datos modificado
head(datos_modificados)
tail(datos_modificados)
```

```r
# Transponer la matriz
datos_transpuestos <- t(data)
head(datos_transpuestos)
```

```r
head(datos_modificados)
tail(datos_modificados)
```

```r

# Sumar los valores de cada edad y crear una nueva fila con las sumas
suma_edades <- colSums(datos_modificados)

# Agregar la fila de sumas al final del conjunto de datos
datos_con_suma <- rbind(datos_modificados, suma_edades)

# Mostrar el conjunto de datos con la nueva fila de sumas
head(datos_con_suma)
tail(datos_con_suma)
```

```r


# Generar una secuencia de 1 a 100
secuencia_edades <- seq(0, 100)

# Reemplazar la primera fila con la secuencia generada
datos_con_suma[1, ] <- secuencia_edades

# Seleccionar las edades y las frecuencias
edades <- as.numeric(datos_con_suma[1, ])
frecuencias <- as.numeric(datos_con_suma[74, ])
# Crear un histograma con barplot

barplot(
  frecuencias,
  names.arg = edades,
  col = "blue",
  border = "black",  # Color del borde de las barras
  density = 40,      # Densidad de las líneas en las barras
  xlim = c(0, 100),  # Límites del eje x
  ylim = c(0, 30000),  # Límites del eje y
  space = 0.2,       # Espacio entre las barras
  col.axis = "darkred",  # Color de las etiquetas de los ejes
  col.lab = "darkblue",  # Color de los títulos de los ejes
  main = "Histograma de Edades",  # Título principal
  sub = "Edad",  # Subtítulo
  cex.main = 1.5,    # Tamaño de la fuente del título principal
  cex.sub = 1.2,     # Tamaño de la fuente del subtítulo
  cex.axis = 0.8     # Tamaño de la fuente de las etiquetas de los ejes
)
```

```r


# Crear un nuevo conjunto de datos con la primera y última fila
nueva_tabla <- rbind(datos_con_suma[1, ], datos_con_suma[nrow(datos_con_suma), ])

# Imprimir el nuevo conjunto de datos
head(nueva_tabla)
```

```r


# Transponer el conjunto de datos
datos_transpuestos <- t(nueva_tabla)

# Crear un nuevo conjunto de datos con columnas renombradas
nuevos_datos <- data.frame(Edad = datos_transpuestos[, 1], `NumeroPersonas` = datos_transpuestos[, 2])

# Imprimir el nuevo conjunto de datos
head(nuevos_datos)
```

```r


# Calcular la media ponderada de edad
media_ponderada_edad <- sum(nuevos_datos$Edad * nuevos_datos$NumeroPersonas) / sum(nuevos_datos$NumeroPersonas)

media_ponderada_edad
```

```r


# Calcular la desviación estándar ponderada de edad
desviacion_estandar_ponderada <- sqrt(
  sum(nuevos_datos$NumeroPersonas * (nuevos_datos$Edad - media_ponderada_edad)^2) / sum(nuevos_datos$NumeroPersonas)
)

desviacion_estandar_ponderada
```

Coeficiente de Variación


```r
coeficiente_variacion <- (desviacion_estandar_ponderada/media_ponderada_edad)
coeficiente_variacion
```

```r


# Crear un nuevo conjunto de datos con las edades ponderadas
edades_ponderadas <- rep(nuevos_datos$Edad, nuevos_datos$NumeroPersonas)

# Crear un data frame con las edades ponderadas
datos_ponderados <- data.frame(EdadesPonderadas = edades_ponderadas)

# Crear el boxplot
boxplot(datos_ponderados$EdadesPonderadas,
        main = "Boxplot de Edades Ponderadas",
        ylab = "Edades Ponderadas")

# Imprimir la media ponderada de edad en el gráfico
abline(h = media_ponderada_edad, col = "red", lty = 2)
```
