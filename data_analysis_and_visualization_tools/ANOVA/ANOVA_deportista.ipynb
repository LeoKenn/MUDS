### <font color='#3d59c6'> PARTE I </font>

Imagina que te llega un conjunto de datos que contiene tres variables categóricas>> A,B y C (con 2 categorías cada una). ¿Podrías plantear una estrategia que recoja los 3 factores realizando un Two-Way ANOVA?

R/: Primero se haría una observación de los datos, para identificar lo que describe su estadístico y ver el comportamiento de los datos. Identificar la media, la mediana, outliers. Identificar si la distribución de los datos tiende a tener una distribución Gaussiana.


Luego, Si las variables continuas tienen las mismas unidades, haría una matriz de covarianza para obtener información de como se relacionan los datos. Luego,  una Matriz de correlación para ver cómo se relacionan las variables entre si y ver si al final, al estar muy correlacionadas, amplifican los efectos entre variables. Esto hace más difícil el poder determinar qué variable o variables están contribuyendo de manera significativa a las diferencias observadas entre los grupos
Luego se comprobarían supuestos de Independencia de las muestras, Normalidad por categorías y Homocedasticidad entre grupos.

Luego se podrían plantear las siguientes hipótesis nulas y alternativas:


Hipótesis nula de efecto de A: No hay diferencias significativas en la variable (dependiente) continua entre las dos categorías de A.


Hipótesis nula de efecto de B: No hay diferencias significativas en la variable continua entre las dos categorías de B.


Hipótesis nula de efecto de C: No hay diferencias significativas en la variable continua entre las dos categorías de C.


Hipótesis nula de interacción entre los factores: No hay una interacción significativa entre A, B y C en el efecto sobre la variable continua.


Luego se realiza el análisis de varianza con los datos de la variable dependiente y las tres variables categóricas. Se calculan los valores F  y el valor-p asociado a los factores. Si el valor de p es menor que el nivel de significancia (puede ser 0.05), se rechaza la hipótesis nula y se concluye que existe un efecto significativo en la variable dependiente.


 ¿y un One-Way ANOVA? ¿Qué consideraciones tendrías en cuenta al diseñar estos análisis?

 R/: Se identificaría la variable continua dependiente y luego se podría comparar por categoría. Por ejemplo se puede crear A1 y A2 en la variable A, B1 y B2 en la variable B, y C1 y C2 en la variable C.

 Luego se Aplica el One-Way ANOVA en el conjunto de datos utilizando la variable dependiente que habíamos seleccionado y las categorías definidas anteriormente. Así, podremos determinar si hay diferencias significativas en las medias de la variable dependiente entre las diferentes categorias. (Entendiendo que el One-Way ANOVA permite determinar si existen diferencias significativas en las medias de la variable continua dependiente entre categorias).
 Si el resultado del One-Way ANOVA es estadísticamente significativo, indicaría que hay al menos un grupo que difiere de los demás en términos de la variable continua dependiente.

## <font color='#3d59c6'>  4. TEST ANOVA </font>

### COMPROVACIÓN DE SUPUESTOS


- **Independencia en las muestras**: Vamos a suponer que las muestras fueron tomadas de forma independiente sin depender de ninguna otra observación.
- **Distribución normal** por categorías
- **Homocedasticidad** entre grupos

#### NORMALIDAD factor sex

```r
variables = c('rcc', 'wcc', 'hc', 'hg', 'bmi')
categorias = c('f', 'm')
resultados <- data.frame(Variable = character(), Cat = character(), Shapiro = numeric(), PValor = numeric(), row.names = NULL)

for (variable in variables) {
  for (cat in categorias) {
    data_subset = subset(data_ais, sex == cat)
    result = shapiro.test(data_subset[, variable])
    resultados = rbind(resultados, data.frame(Variable = variable, Cat = cat, Shapiro = result$statistic, PValor = round(result$p.value, 3)))
  }
}

print(resultados)
```

Todos los valores por encima de p 0.05 por lo tanto se acepta la hipotesis nula de normalidad. Excepto en rcc female, donde el valor es 0.018 por debajo de 0.05. Por lo tanto hay evidencia suficiente para rechazar la hipótesis nula de normalidad.

```r
female_data <- data_ais[data_ais$sex == "f", ]

#Histograma solo para 'f'
ggplot(female_data, aes(x = rcc, fill = sex)) +
  geom_histogram(binwidth = 0.2, position = "identity", alpha = 0.7) +
  labs(title = "Histograma de 'rcc' para 'female'",
       x = "Valor de 'rcc'",
       y = "Frecuencia") +
  scale_fill_manual(values = c("#1f78b4")) +
  theme_minimal()
```

#### NORMALIDAD factor Sport

```r
variables = c('rcc', 'wcc', 'hc', 'hg', 'bmi')
categorias = c('B_Ball', 'Row', 'Swim', 'T_400m', 'Tennis')
resultados <- data.frame(Variable = character(), Cat = character(), Shapiro = numeric(), PValor = numeric(), row.names = NULL)

for (variable in variables) {
  for (cat in categorias) {
    data_subset = subset(data_ais, sport == cat)
    result = shapiro.test(data_subset[, variable])
    resultados = rbind(resultados, data.frame(Variable = variable, Cat = cat, Shapiro = result$statistic, PValor = round(result$p.value, 3)))
  }
}

print(resultados)
```

Todos los valores por encima de p 0.05 por lo tanto se acepta la hipotesis nula de normalidad. Destaca Rcc swim con un valor p cercano a 0.05

#### HOMOCEDASTICIDAD factor sex

- $H_{0}$: las muestras presentan varianzas iguales
- $H_{1}$: las muestras presentan varianzas distintas

```r
variables = c('rcc', 'wcc', 'hc', 'hg', 'bmi')
resultados_bartlett = data.frame(Variable = character(), Estadistico = numeric(), PValor = numeric(), row.names = NULL)

for (variable in variables) {
  resultado_bartlett = bartlett.test(get(variable) ~ sex, data = data_ais)
  resultados_bartlett = rbind(resultados_bartlett, data.frame(Variable = variable, Estadistico = resultado_bartlett$statistic, PValor = resultado_bartlett$p.value))
}

print(resultados_bartlett)
```

#### HOMOCEDASTICIDAD factor sport

```r
variables = c('rcc', 'wcc', 'hc', 'hg', 'bmi')
resultados_bartlett = data.frame(Variable = character(), Estadistico = numeric(), PValor = numeric(), row.names = NULL)

for (variable in variables) {
  resultado_bartlett = bartlett.test(get(variable) ~ sport, data = data_ais)
  resultados_bartlett = rbind(resultados_bartlett, data.frame(Variable = variable, Estadistico = resultado_bartlett$statistic, PValor = resultado_bartlett$p.value))
}

print(resultados_bartlett)
```

Todos los valores por encima de p 0.05, por lo tanto se acepta la Hipótesis alternativa, las muestras presentan varianzas distintas

### <font color='#3d59c6'>  4.1 One-Way ANOVA </font>

Análisis de Varianza (ANOVA): técnica estadística que se utiliza para evaluar si existen diferencias significativas entre las medias de diferentes grupos, considerando la variabilidad tanto dentro de cada grupo como entre los grupos

- $H_{0}$: No existen diferencias significativas entre las medias de los grupos
- $H_{1}$: Existen diferencian significativas entre las medias de los grupos

A partir de este punto vamos a centrarnos en si existe diferencia entre los grupos definidos por nuestro factor 'sport' y las variables dependientes continuas "wcc" y "hc" que presentaron un bajo nivel de correlación cercano a 0

```r
ggplot(data_ais, aes(sport, wcc)) +
  geom_violin(color = NA,aes(fill = factor(sport))) +
  theme(legend.position = 'none') +
  geom_boxplot(width = 0.1) +
  scale_fill_manual(values = c('B_Ball' = '#ffb6c1', 'Row' = '#98fb98', 'Swim' = '#fdfd96', 'T_400m' = '#87ceeb', 'Tennis' = '#D8BFD8'))
```

```r
ggplot(data_ais, aes(sport, hc)) +
  geom_violin(color = NA,aes(fill = factor(sport))) +
  theme(legend.position = 'none') +
  geom_boxplot(width = 0.1) +
  scale_fill_manual(values = c('B_Ball' = '#ffb6c1', 'Row' = '#98fb98', 'Swim' = '#fdfd96', 'T_400m' = '#87ceeb', 'Tennis' = '#D8BFD8'))
```

```r
anova = aov(wcc ~ sport, data = data_ais)
summary(anova)
```

```r
resultados_anova = data.frame(Variable = character(), F_Value = numeric(), P_Value = numeric(), row.names = NULL)
variables = c('wcc', 'hc')
for (variable in variables) {
  formula = as.formula(paste(variable, '~ sport'))
  anova_result = aov(formula, data = data_ais)

  # Extraer F value y p-value
  f_value = summary(anova_result)[[1]][['F value']][1]
  p_value = summary(anova_result)[[1]][['Pr(>F)']][1]
 resultados_anova = rbind(resultados_anova, data.frame(Variable = variable, F_Value = f_value, P_Value = p_value))
}

# Muestra los resultados de ANOVA
print(resultados_anova)
```

Este resultado nos indica con un p-valor > 0.05 que debemos aceptar nuestra hipótesis nula y por lo tanto no hay diferencia significativa entre grupos.
No hay diferencias significativas entre las medias de los grupos en términos de la variable "hc" y "wcc".

El tipo de deporte no es determinante para los valores medios de hc y wcc
