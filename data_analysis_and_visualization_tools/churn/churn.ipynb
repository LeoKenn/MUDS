### <font color='#3d59c6'> Desarrollo </font>

Utilizaría la siguiente estrategia para desarrollar un modelo predictivo:

Análisis exploratorio de datos: Realizaría un análisis detallado de las variables para conocer si son continuas o categóricas, identificar la variable objetivo (Buscaría una variable del estilo "consumo" en km/litro), identificaría outliers y valores nulos, correlaciones y distribuciones. Con esto se podrá entender mejor los datos y la relación entre ellos.

Preprocesamiento de datos: Una vez analizados los datos, procedería a limpiar el dataset, identificando si hacen falta muchos datos o si hay outliers. Si hay alguna variable que no deberíamos tomar en cuenta debido a la falta de información. Buscaría hacer una codificación de las variables categóricas, por ejemplo, fin de semana "1" y entre semana "0", para asegurarnos que los datos están en un formato adecuado para el modelado.

Observaría si fuera necesario hacer una estandarización o normalización de los datos. Algunos en km otros en grados Celsius. Observar que los órdenes de magnitud no sean muy dispares.

Selección de variables: Una vez entendido el dominio y los datos procedería a seleccionar un grupo de variables que permitan generar el modelo. Demasiadas variables podrían generar ruido y no ser beneficiosas para el modelo.

Iniciaría con el análisis de correlación, identificando aquellas variables que están altamente correlacionadas entre sí. También observando aquellas que tienen alta relación con la variable objetivo.

De estas 31, supongamos nos quedamos con unas 20 que no están tan correlacionadas y que pueden ser útiles para generar el modelo.

Luego procedería a hacer un análisis PCA, para identificar las componentes principales. De estas componentes buscaría cuantas son capaces de explicar la mayor cantidad de variabilidad de los datos (> 80 % por ejemplo). De estas componentes, vería en valor absoluto, aquellas que tienen mayor peso en las componentes principales. De estas seleccionaría, dependiendo del caso, las 3 más importantes de las primeras 2 componentes.

Luego de tener unas 6 variables, procedería a hacer un análisis ANOVA, para ver si hay relación entre las variables categóricas y la variable objetivo.

A priori, vería que las 4 variables categóricas que nos plantean, tienen una influencia directa en la media de la variable objetivo.

Nos quedaríamos entonces, por poner un ejemplo, con 6 variables continuas y 4 categóricas. Procedería ahora a crear un modelo de regresión lineal polinómico, por ejemplo.

División de datos: Separaría el conjunto de datos en conjuntos de entrenamiento y prueba. El conjunto de entrenamiento se utilizaría para entrenar el modelo y el conjunto de prueba se utilizaría para evaluar el rendimiento.

Entrenamiento del modelo: Utilizaría el conjunto de entrenamiento para ajustar los parámetros del modelo y encontrar la mejor configuración.

Calidad del modelo: Utilizaría el conjunto de prueba para evaluar el rendimiento del modelo, con métricas como el coeficiente de determinación (R^2) y F de fisher. Buscando valores de R^2 cercanos a 1, F-statistic alto y "pvalue<0.05".

Validación del modelo: Tomaría los residuos y analizaría se tienen un comportamiento normal con la prueba Shapiro-Wilk, vería la Homocedasticidad con Breusch-Pagan test, Autocorrelación con Durbin-Watson y detección outliers con la distancia de Cook)

Luego haría una valoración del modelo utilizando las métricas  del Mean Absolute error (MAE) y Root Mean Squared Error (RMSE). Este último me permitiría identificar de manera más directa que tan acertado es el modelo, ya que si fuera una regresión lineal simple tendría las mismas unidades de la variable objetivo.

Elección del modelo: Si el modelo de regresión lineal no fuera el más apropiado, buscaría utilizar otros modelos como árboles de decisión, random Forest o modelos de redes neuronales para ver si tienen una precisión más precisa sin que lleven al overfitting.

```r
# Generamos datasets filtrando por la clase
data_churn_num_1 <- data_churn_num[data_churn_num$is_churned == 1,]
data_churn_num_0 <- data_churn_num[data_churn_num$is_churned == 0,]
```

```r
# Calcular la media para churn = 1
means_churn_1 <- colMeans(data_churn_num_1, na.rm = TRUE)

# Calcular la media para churn = 0
means_churn_0 <- colMeans(data_churn_num_0, na.rm = TRUE)

# Combinar las medias en un dataframe y formatear para evitar notación científica
means_table <- cbind(Churn_1 = format(means_churn_1, scientific = FALSE, digits = 3),
                     Churn_0 = format(means_churn_0, scientific = FALSE, digits = 3))

# Convertir a data frame
means_table <- as.data.frame(means_table)

# Mostrar la tabla
means_table
```

```r
# Función para calcular la moda
Mode <- function(x) {
  ux <- unique(x)
  ux[which.max(tabulate(match(x, ux)))]
}

# Calcular la moda para churn = 1
modes_churn_1 <- apply(data_churn_num_1, 2, Mode)

# Calcular la moda para churn = 0
modes_churn_0 <- apply(data_churn_num_0, 2, Mode)

# Combinar las modas en un dataframe
modes_table <- data.frame(Churn_1 = modes_churn_1, Churn_0 = modes_churn_0)

# Mostrar la tabla
modes_table
```

```r
# Calcular la mediana para churn = 1
medians_churn_1 <- apply(data_churn_num_1, 2, median, na.rm = TRUE)

# Calcular la mediana para churn = 0
medians_churn_0 <- apply(data_churn_num_0, 2, median, na.rm = TRUE)

# Combinar las medianas en un dataframe
medians_table <- data.frame(Churn_1 = medians_churn_1, Churn_0 = medians_churn_0)

# Mostrar la tabla
medians_table
```

```r
# Calcular la desviación estándar para churn = 1
sds_churn_1 <- apply(data_churn_num_1, 2, sd, na.rm = TRUE)

# Calcular la desviación estándar para churn = 0
sds_churn_0 <- apply(data_churn_num_0, 2, sd, na.rm = TRUE)

# Formatear las desviaciones estándar para evitar la notación científica y mostrar solo 3 cifras significativas
sds_churn_1_formatted <- format(sds_churn_1, scientific = FALSE, digits = 3)
sds_churn_0_formatted <- format(sds_churn_0, scientific = FALSE, digits = 3)

# Combinar las desviaciones estándar en un dataframe
sds_table <- data.frame(Churn_1 = sds_churn_1_formatted, Churn_0 = sds_churn_0_formatted)

# Mostrar la tabla
sds_table
```

Observamos que tanto la media como la mediana tiene valores muy similares en ambos grupos. Somo la desviación estandar muestra cierta diferencia en algunas variables como "payments_completed_amount_first_7days", "coins_redeemed_first_7days"

#### <font color='#3d59c6'> REALIZACIÓN IGA
 </font>

Con tal de ver la información mutua entre el target y las covariables se aplican los tres algoritmos detallados a continuación:

- "infogain": ${H(Class)} + H(Attribute) − H(Class, Attribute)$

- "gainratio" : $H(Class) + H(Attribute) − H(Class, Attribute) / H(Attribute) $

- "symuncert" : $2 * (H(Class) + H(Attribute) − H(Class, Attribute)) / (H(Attribute) + H(Class)) $

```r
options(repr.plot.width = 8, repr.plot.height = 8)

# Quita la variable user_id del análisis
ig <- information_gain(formula = is_churned ~ . - user_id,
                       data = data_churn_num,
                       type = 'infogain')

# Crea el gráfico de barras horizontales
ggplot(ig, aes(x = reorder(attributes, +importance), y = importance)) +
  geom_bar(stat = "identity", fill = "#990000", color = "#990000") +
  geom_text(aes(label = round(importance, 3)), color = "black", size = 6) +  # Agrega etiquetas de valores
  coord_flip() +
  labs(x = "Attributes", y = "Importance") +
  theme(axis.text.y = element_text(size = 12))
```

### <font color='#3d59c6'> Sin Balanceo </font>


Como habiamos visto anteriormente, la variable objetivo esta desbalanceada y esto causa que el modelo no prediga bien la clase minoritaria.

#### <font color='#3d59c6'>Fase de Entrenamiento </font>


```r
set.seed(5)
inTrain = createDataPartition(y = data_churn_num$is_churned
                                  , p = .80
                                  , list = FALSE
                                  , times = 1)
data_churn_num.train = data_churn_num[inTrain,]
data_churn_num.test = data_churn_num[-inTrain,]

ytrain = data_churn_num$is_churned[inTrain]
ytest = data_churn_num$is_churned[-inTrain]
```

##### <font color='#3d59c6'>Modelo de 1 variable </font>


```r
### Generamos un primer modelo unicamente con la variable payments_completed
mrl0 = glm(formula = is_churned ~ payments_completed
          , data = data_churn_num.train
          , family = "binomial")
summary(mrl0)
```

```r
### Generamos las predicciones de nuestro modelo
mrl0_prob = predict(mrl0, type = 'response')
head(mrl0_prob)
```

```r
### Vemos la predicción para cada instancia
mrl0_pred = ifelse(mrl0_prob > 0.5, 1, 0)
head(mrl0_pred)
```

```r
confusionMatrix(as.factor(mrl0_pred), as.factor(data_churn_num.train$is_churned))
```


* Verdaderos positivos (TP): 0
* Falsos positivos (FP): 0
* Verdaderos negativos (TN): 57995
* Falsos negativos (FN): 23090

Observamos que este primer modelo no está detectando los casos positivos.

##### <font color='#3d59c6'>Modelo de multiples variables </font>


```r
### Generamos un primer modelo unicamente con la variable payments_completed
mrl1 = glm(formula = is_churned ~ payments_completed +  first_payment_amount + coins_redeemed_first_7days + is_referral
          , data = data_churn_num.train
          , family = "binomial")
summary(mrl1)
```

### <font color='#3d59c6'> Con Balanceo </font>


Se empiezan a observar clasificaciones de la variable objetivo

#### <font color='#3d59c6'>Fase de Entrenamiento </font>


```r
# Contar el número de observaciones para cada clase
count_churned_0 <- sum(data_churn_num$is_churned == 0)
count_churned_1 <- sum(data_churn_num$is_churned == 1)

# Determinar la clase con menos observaciones
minority_class <- ifelse(count_churned_0 < count_churned_1, 0, 1)
majority_class <- ifelse(minority_class == 0, 1, 0)

# Submuestrear la clase mayoritaria
data_majority <- data_churn_num[data_churn_num$is_churned == majority_class, ]
sampled_indices <- sample(1:nrow(data_majority), count_churned_1)

# Crear el conjunto de datos balanceado
data_balanced <- rbind(data_majority[sampled_indices, ], data_churn_num[data_churn_num$is_churned == minority_class, ])

# Dividir los datos en conjuntos de entrenamiento y prueba
set.seed(6)
inTrain <- createDataPartition(y = data_balanced$is_churned, p = .80, list = FALSE)
data_train_balanced <- data_balanced[inTrain, ]
data_test_balanced <- data_balanced[-inTrain, ]

# Separar las etiquetas de clase de los conjuntos de entrenamiento y prueba balanceados
ytrain_balanced <- data_train_balanced$is_churned
ytest_balanced <- data_test_balanced$is_churned
```

```r
### Generamos un primer modelo unicamente con la variable payments_completed
mrl2 = glm(formula = is_churned ~ payments_completed
          , data = data_train_balanced
          , family = "binomial")
summary(mrl2)
```

```r
### Generamos las predicciones de nuestro modelo
mrl2_prob = predict(mrl2, type = 'response')
head(mrl2_prob)
```
